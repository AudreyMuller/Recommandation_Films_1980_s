```python
df_ourmovies = pd.read_csv('/content/drive/MyDrive/2209-Data/Projets/Projet2/MMA/00_CSV_files/00_DATABASE_FINAL.csv', sep=',')
```

```python
df_ourmovies.drop(df_ourmovies.filter(regex="Unname"),axis=1, inplace=True)
```

## Merge the Directors with ourmovies Database

```python
df_ourmoviesdirectors = df_ourmovies.merge(df_directors, on='tconst')
```

```python
df_ourmoviesdirectors.head()
```

## delete columns we do not need

```python
df_ourmoviesdirectors.drop(['genres_y','primaryTitle_y','startYear_y','runtimeMinutes_y','averageRating_y','Weight'],axis=1, inplace=True)
```

## rename some columns

```python
df_ourmoviesdirectors.rename(columns = {'primaryTitle_x':'primaryTitle', 'startYear_x':'startYear','runtimeMinutes_x':'runtimeMinutes','genres_x':'genres','averageRating_x':'averageRating'}, inplace = True)
```
