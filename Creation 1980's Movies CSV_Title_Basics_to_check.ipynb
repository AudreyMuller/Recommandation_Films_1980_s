## READ CSV

```python
df=pd.read_csv('https://datasets.imdbws.com/title.basics.tsv.gz',sep='\t')
```

#REPLACE \N and DROP LINES \N

```python
# Remplace the \N by None in 'runtimeMinutes'
df.replace({r'\N': None},inplace =True)
```

```python
#define tconst as Index
df.set_index('tconst',inplace = True)

```

```python
#drop \N in some columns
df.dropna(how = 'any',subset = ['primaryTitle','originalTitle','isAdult','startYear','genres'], inplace=True)
```

#CONVERT columns to the good type

```python
#convert Séries in the good type
df['runtimeMinutes']= pd.to_numeric(df['runtimeMinutes'],'coerce','integer')
#.fillna(0).astype(np.int64)
```

```python
#convert 'startYear' in integer
df=df.astype({'startYear':np.int64,'isAdult':np.bool_})
```

#DROP LINES to keep 1980's Movies

```python
#Drop IsAdult = True
df=df[df['isAdult']==False]
```

```python
# keep only the movies in 1980's
df=df[(df['titleType']=='movie') & (df['startYear']>=1980) & (df['startYear']<1990)]
```

#CONVERT the last column in the good type

```python
#convert 'genres' en list  --> .split(',')
#with the apply which is long, prefer to run after the filter on 1980's movies
df['genres'] = df['genres'].apply(lambda x: x.split(','))
```

#EXPLODE genres to filter by genre

```python
df_explode=df.explode('genres')
```

```python
df_final_explode=df_explode[df_explode['genres'].isin(['Drama','Comedy','Action','Romance','Crime','Thriller','Adventure','Horror','Family','Fantasy','War','Mystery','Sci-Fi','Animation'])]
```

#FILTER DataFrame to keep tconst in df_final_explode

```python
df_final=df[df.index.isin(list(df_final_explode.index))]
```
